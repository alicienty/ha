```python
oleg_text = oleg_text.lower()
oleg_text[500:1000]
```

```python
oleg_text = re.sub(r'[^а-яА-ЯёЁ\s\.\?!]', '', oleg_text)
oleg_text[500:1000]
```

```python
sentences = re.split(r'[\.\?!]', oleg_text)
sentences[1:3]
```

```python
sentences = [sentence.split() for sentence in sentences if sentence]
sentences[3]
```

```python
co_occurrence = defaultdict(lambda: defaultdict(int))
```

```python
def update_co_occurrence(sentence, window_size=1):
    words = ["<START>"] + sentence + ["<END>"]
    for i, word in enumerate(words):
        for j in range(max(0, i - window_size), min(len(words), i + window_size + 1)):
            if i != j:
                co_occurrence[word][words[j]] += 1

for sentence in sentences:
    update_co_occurrence(sentence)

co_occurrence['позже']
```

```python
tokens = list(co_occurrence.keys())
tokens[:10]
```

```python
co_occurrence_df = pd.DataFrame(co_occurrence).fillna(0).astype(int).reindex(tokens, columns=tokens, fill_value=0)
co_occurrence_df
```

```python
oleg_co_occurence = co_occurrence['олег']
```

```python
max_key = max(oleg_co_occurence, key=oleg_co_occurence.get)
```

```python
max_key
print (f'Чаще всего "олег" встречается с {max_key}.')
```

```python
#посмотрим на топ-10

top_10 = sorted(oleg_co_occurence.items(), key=lambda x: x[1], reverse=True)[:10]

print(top_10)
```
